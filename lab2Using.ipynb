```python
from tensorflow.keras.models import load_model
from IPython.display import Image
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
classes = ['футболка', 'брюки', 'светр', 'плаття', 'пальто', 'туфлі', 'сорочка', 'кросівки', 'сумка', 'чоботи']
model = load_model('fashion_mnist.h5')
```

```python
#завантаження картинки
img_list = ["image/img_1.png", "image/img_2.png", "image/img_3.png", "image/img_4.png", "image/img_5.png"]
for imagee in img_list:
    img = image.load_img(imagee, target_size=(28, 28), color_mode = "grayscale")


    # перетворюємо зображення в масив
    x = image.img_to_array(img)
    # змінюємо форму масиву в плаский вектор
    x = x.reshape(1, 784)
    # інвертуємо зображення
    x = 255 - x
    # нормалізуємо зображення
    x = x/255
    #розпізнавання
    prediction = model.predict(x)

    prediction = np.argmax(prediction)
    # картинка
    plt.imshow(img.convert('RGBA'))
    plt.title(classes[prediction])
    plt.show()
```
